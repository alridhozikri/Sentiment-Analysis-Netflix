```python
from google_play_scraper import Sort, reviews
import pandas as pd

app_id = 'com.netflix.mediaclient'

result, _ = reviews(
    app_id,
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=4000
)
```

```python
df = pd.DataFrame(result)
df_save = df[['content', 'score']]
```

```python
df_save = df_save.drop_duplicates(subset=['content'])
df_save = df_save.dropna()
```

```python
df_save.to_csv('dataset_scraping_netflix.csv', index=False)
print(f"Dataset berhasil disimpan! {len(df_save)}")
```
